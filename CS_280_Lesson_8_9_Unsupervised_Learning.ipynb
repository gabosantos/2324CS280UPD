# Concepts

Although most of the applications of ML today are based on supervised learning, the vast majority of the available data is unlabeled: we have the input features X, but we do not have the labels y.

Yann LeCun famously said:
> If intelligence was a cake, unsupervised learning would be the cake, supervised learning would be the icing on the cake, and reinforcement learning would be the cherry on the cake.

In other words, there is a huge potential in unsupervised learning that we have only barely started to sink our teeth into [HML].

## Unsupervised Learning

<font color="#9980FA">Unsupervised learning</font> serves a myriad of purposes, which include [HML]:

1. <font color="#9980FA">**Dimensionality Reduction**</font>:
The goal is to reduce the number of features significantly, transforming intractable problems into manageable ones.

2. <font color="#9980FA">**Clustering**</font>:
The goal is to group similar instances together into clusters, usefule for data analysis, customer segmentation, recommender systems, search engines, image segmentation, and more.

3. <font color="#9980FA">**Anomaly detection**</font>:
The objective is to learn what “normal” data looks like, and then use that to detect abnormal instances, such as defective items on a production line or a new trend in a time series.

## Dimensionality Reduction

Many ML problems involve thousands or even millions of features for each training instance. Not only do all these features make training extremely slow, but they can also make it much harder to find a good solution, as we will see. This problem is often referred to as the <font color="#9980FA">curse of dimensionality</font> [HML].

We are so used to living in three dimensions that our intuition fails us when we try to imagine a high-dimensional space. Even a basic 4D hypercube is incredibly hard to picture in our minds, let alone a 200-dimensional ellipsoid bent in a 1,000-dimensional space [HML].

<br>

<center><small>Curse of dimensionality [HML]</small></center>

<br>

It turns out that many things behave very differently in high-dimensional space [HML]. For example, expanding the dimension from 2D to 3D would create more space, and increase the chance of points being sparse (on the corners) [[Di Salvo](https://medium.com/geekculture/curse-of-dimensionality-e97ba916cb8f)].

<br>

There are two primary approaches to reducing dimensionality: Projection and Manifold Learning [HML].

### Projection
<font color="#9980FA">Projection</font> is similar to the kernel trick, which involves transforming data. But instead of moving to a higher dimension, it aims to lower the dimensionality of the data.

In the example below, we can project every training instance perpendicularly onto a lower-dimensional subspace (as represented by the short lines connecting the instances to the plane). This reduces the data set's dimensionality from 3D to 2D [HML].

### Manifold Learning

However, projection is not always the best approach to dimensionality reduction. In many cases the subspace may twist and turn, such as in the famous Swiss roll toy data set shown below [HML].

<br>

<table width="100%"><tr><td align="center" bgcolor="white"><img src="https://drive.google.com/uc?export=view&id=1xzxmLyPcsgcTVkH19vVlhb3h02lb5s5R" height="360"/></td></tr></table>
<center><small>Swiss roll data [HML]</small></center>

<br>

Simply projecting onto a plane (e.g., by dropping x3) would squash different layers of the Swiss roll together, as shown on the left side of the figure below. What you really want is to unroll the Swiss roll to obtain the 2D dataset on the right side of figure [HML].

<br>

<table width="100%"><tr><td align="center" bgcolor="white"><img src="https://drive.google.com/uc?export=view&id=1hHdzM38wAhT2PvzVMfh0Fk9hy4c3D7-Q" height="360"/></td></tr></table>
<center><small>Projecting vs. Unrolling data [HML]</small></center>

<br>

This technique is called <font color="#9980FA">manifold learning</font>. Put simply, a $d$-dimensional manifold is a $d$-dimensional shape that can be bent and twisted in a higher-dimensional space.

For example, think about the MNIST dataset: all handwritten digit images have some similarities. They are made of connected lines, the borders are white, and they are more or less centered. If you randomly generated images, only a ridiculously tiny fraction of them would look like handwritten digits. In other words, the degrees of freedom available to you if you try to create a digit image are dramatically lower than the degrees of freedom you would have if you were allowed to generate any image you wanted. These constraints tend to squeeze the dataset into a lower-dimensional manifold [HML].

<br>

<br>

Note that manifold learning assumes that the task at hand (e.g., classification or regression) will be simpler if expressed in the lower-dimensional space of the manifold. But this this implicit assumption does not always hold. Sometimes, unrolling the manifold results to a more complex decision boundary.

## Principal Component Analysis

<font color="#9980FA">Principal Component Analysis (PCA)</font> is by far the most popular dimensionality reduction algorithm. First it identifies the hyperplane that lies closest to the data, and then it projects the data onto it [HML].

The first step in PCA is finding the hyperplane, also known as decorrelation. This hyperplane doesn't alter the dimensionality of the data; instead, it is chosen to maximize the variance captured in the original dataset along this specific direction.

Shown below are three potential hyperplanes (lines): solid, dashed, dotted. It seems reasonable to select the solid line since it preserves the maximum amount of variance, and thus will most likely lose less information than the other projections [HML].

<br>

<table width="100%"><tr><td align="center" bgcolor="white"><img src="https://drive.google.com/uc?export=view&id=1rpPgxYdj1pQ2LMw_ScPjUgklpd8_IYyI" height="360"/></td></tr></table>
<center><small>Finding the best hyperplane in PCA [HML]</small></center>

Another way to justify this choice is that it is the axis that minimizes the mean squared distance between the original dataset and its projection onto that axis. This is the rather simple idea behind PCA [HML].

PCA identifies the axis that accounts for the largest amount of variance in the data. In the figure above, it is the solid line.

It also finds a second axis, orthogonal to the first one, that accounts for the largest amount of remaining variance. In the figure above, it is the dotted line.

If it were a higher-dimensional data set, PCA would also find a third axis, orthogonal to both previous axes, and a fourth, and so on—as many axes as the number of dimensions in the dataset.

To find the <font color="#9980FA">principal components</font> of a training set, we can use a standard matrix factorization technique called <font color="#9980FA">Singular Value Decomposition (SVD)</font> that can decompose the training set matrix $\mathbf{X}$ into the matrix multiplication of three matrices $\mathbf{U} \mathbf{\Sigma} \mathbf{V}^T$, where $\mathbf{V}$ contains the unit vectors that define all the principal components that we are looking for [HML].

<br>

\begin{equation}
\mathbf{X} = \begin{bmatrix}
  \vert & \vert & & \vert \\
  u_1 & u_2 & \dots & u_m \\
  \vert & \vert & & \vert
\end{bmatrix}
\begin{bmatrix}
  \sigma_1 & 0 & \dots & 0 \\
  0 & \sigma_2 & \dots & 0 \\
  \vdots & \vdots & \ddots & \vdots \\
  0 & 0 & \dots & \sigma_n
\end{bmatrix}
\begin{bmatrix}
  - & v_1^T & - \\
  - & v_2^T & - \\
  & \vdots & \\
  - & v_n^T & -
\end{bmatrix}
\end{equation}

<br>


Another way of finding the principal components is by computing the covariance matrix and performing eigenvalue decomposition.

Once we have identified all the principal components, the next step is to reduce the dimensionality of the dataset down to $d$-dimensions by projecting it onto the hyperplane defined by the first $d$ principal components [HML]:

\begin{equation}
\mathbf{X}_d = \mathbf{X} \mathbf{W}_d
\end{equation}

where $\mathbf{W}_d$ is the matrix containing the first $d$ columns of $\mathbf{V}$.

## Clustering

Clustering algorithms seek to learn, from the properties of the data, an optimal division or discrete labeling of groups of points.

Many clustering algorithms are available in Scikit-Learn and elsewhere, but perhaps the simplest to understand is an algorithm known as k-means clustering, which is implemented in `sklearn.cluster.KMeans` [PDSH].

### K-Means Clustering


The <font color="#9980FA">*k*-means algorithm</font> searches for a predetermined number of clusters within an unlabeled multidimensional dataset.
It accomplishes this using a simple conception of what the optimal clustering looks like:

- The *cluster center* is the arithmetic mean of all the points belonging to the cluster.
- Each point is closer to its own cluster center than to other cluster centers.

Those two assumptions are the basis of the *k*-means model [PDSH].


*k*-means clustering algorithm works as follows:

* <font color="#9980FA">**Initialization**</font>: Randomly choose $k$ initial cluster centroids.

* <font color="#9980FA">**Assignment**</font>: Assign each data point to the nearest centroid, forming k clusters.

* <font color="#9980FA">**Update**</font>: Recalculate the centroids as the mean of the data points in each cluster.

* <font color="#9980FA">**Repeat**</font>: Iterate the assignment and update steps until convergence or a specified number of iterations.

<br>

Although the algorithm is guaranteed to converge, it may not converge to the right solution (i.e., it may converge to a local optimum): whether it does or not depends on the centroid initialization.

The main challenge with using *k*-means clustering is that the number of clusters should be known at the start. But in general, it will not be so easy to know how to set $k$, and the result might be quite bad if you set it to the wrong value [HML].

We can address this problem by using the elbow method. In the elbow method, we run k-means through a series of candidate k values, and measure the silhouette score.

The tricky part of using *k*-means clustering is figuring out how many clusters (denoted by $k$) you need from the start. In general, it will not be so easy to know how to set $k$, and the result might be quite bad if you set it to the wrong value [HML].


To tackle this issue, there's a handy method called the elbow method. With the <font color="#9980FA">elbow method</font>, we try out different
$k$ values, and check the silhouette score for the number of clusters that works best.

The <font color="#9980FA">silhoutte score</font> (or coefficient) is equal to [HML]:

\begin{equation}
\text{Silhouette Score} = \frac{1}{N} \sum_{i=1}^{N} \frac{b_i - a_i}{\max\{a_i, b_i\}}
\end{equation}

Where:
* $a$ is the mean distance to the other instances in the same cluster (i.e., the mean intra-cluster distance)
* $b$ is the mean nearest-cluster distance (i.e., the mean distance to the instances of the next closest cluster, defined as the one that minimizes b, excluding the instance's own cluster).

<br>

<table width="100%"><tr><td align="center" bgcolor="white"><img src="https://drive.google.com/uc?export=view&id=18xjg_8dk878hg5df99gl4A8l9KlvOlaN" height="360"/></td></tr></table>
<center><small>Elbow method with Silhouette Score [HML]</small></center>


<br>

The silhouette score can vary between –1 and +1. A score close to +1 means that the instance is well inside its own cluster and far from other clusters, while a score close to 0 means that it is close to a cluster boundary, and finally a score close to –1 means that the instance may have been assigned to the wrong cluster [HML].

### Other Clustering Methods


There are other clustering methods aside from *k*-means, such as hierarchical clustering, DBSCAN (Density-Based Spatial Clustering of Applications with Noise), and Gaussian Mixture Models (GMM).

<font color="#9980FA">Hierarchical clustering</font> builds a hierarchy of clusters, organizing them into a tree-like structure that can be visually represented as a dendrogram.

<font color="#9980FA">DBSCAN</font> identifies clusters based on the density of data points, making it robust against noise and capable of discovering clusters of arbitrary shapes.

<font color="#9980FA">Gaussian Mixture Models</font>, on the other hand, assume that the data is generated from a mixture of several Gaussian distributions, allowing for more flexible cluster shapes.

If you're interested in exploring these clustering techniques, you can refer to the `sklearn.cluster` module for implementation details and examples.

## PCA

Let's take a quick look at the application of PCA to the MNIST digits dataset [PDSH]:

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

### Reducing dimension

Recall that the digits dataset consists of 8 × 8–pixel images, meaning that they are 64-dimensional. To gain some intuition into the relationships between these points, we can use PCA to project them into a more manageable number of dimensions, say two [PDSH]:

```python
from sklearn.decomposition import PCA
pca = PCA(2)  # project from 64 to 2 dimensions
projected = pca.fit_transform(digits.data)
print(digits.data.shape)
print(projected.shape)
```

We can now plot the first two principal components of each point to learn about the data, as seen in the following figure [PDSH]:

```python
plt.scatter(projected[:, 0], projected[:, 1],
            c=digits.target, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('rainbow', 10))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar()
plt.show()
```

From here we can see some good separations and clustering already, specifically with 0, 1, 4 and 6. But we only retained 2 principal components, and that might be too many for this high-dimensional image recognition task.

### Choosing number of components

A vital part of using PCA in practice is the ability to estimate how many components are needed to describe the data.

This can be determined by looking at the cumulative *explained variance ratio* as a function of the number of components [PDSH]:

```python
pca = PCA().fit(digits.data)
plt.plot(np.cumsum(pca.explained_variance_ratio_), 'o-')
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

This curve quantifies how much of the total, 64-dimensional variance is contained within the first $N$ components.
For example, we see that with the digits data the first 10 components contain approximately 75% of the variance, while you need around 50 components to describe close to 100% of the variance.

This tells us that our 2-dimensional projection loses a lot of information (as measured by the explained variance) and that we'd need about 20 components to retain 90% of the variance.  Looking at this plot for a high-dimensional dataset can help you understand the level of redundancy present in its features [PDSH].

## K-means clustering

Let's try *k*-means on an artificial 2D dataset containing four distinct blobs. To emphasize that this is an unsupervised algorithm, we will leave the labels out of the visualization [PDSH]:

```python
from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

### Automated clustering

By eye, it is relatively easy to pick out the four clusters.
The *k*-means algorithm does this automatically, and in Scikit-Learn uses the typical estimator API [PSDH]:

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, n_init='auto')
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200);
```

Again, the challenge with *k*-means is that we must tell it how many clusters you expect: it cannot learn the number of clusters from the data.

For example, if we ask the algorithm to identify six clusters, it will happily proceed and find the best six clusters, even though we know that this is incorrect [PDSH]:

```python
labels = KMeans(6, random_state=0, n_init='auto').fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis');
```

### Choosing k

To find the optimal value of K via silhouette scores and elbow method [[Banerji](https://www.analyticsvidhya.com/blog/2021/05/k-mean-getting-the-optimal-number-of-clusters/)]:

```python
from sklearn.metrics import silhouette_score

K = range(2,9)
silhouette_avg = []

for k in K:
  kmeans = KMeans(n_clusters=k, n_init='auto')
  kmeans.fit(X)
  cluster_labels = kmeans.labels_
  silhouette_avg.append(silhouette_score(X, cluster_labels))

# Plot silhouette scores
plt.plot(K, silhouette_avg, 'o-')
plt.xlabel('Values of K')
plt.ylabel('Silhouette score')
plt.title('Silhouette score analysis')
plt.show()
```

### Non-linear clusters

The fundamental model assumptions of *k*-means (points will be closer to their own cluster center than to others) means that the algorithm will often be ineffective if the clusters have complicated geometries.

In particular, the boundaries between *k*-means clusters will always be linear, which means that it will fail for more complicated boundaries [PDSH].

```python
from sklearn.datasets import make_moons
X, y = make_moons(200, noise=.05, random_state=0)

labels = KMeans(2, n_init='auto', random_state=0).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis');
```

This situation is reminiscent of our discussion in Support Vector Machines, where we used a kernel transformation to project the data into a higher dimension where a linear separation is possible.

We might imagine using the same trick to allow *k*-means to discover non-linear boundaries.

One version of this kernelized *k*-means is implemented in Scikit-Learn within the ``SpectralClustering`` estimator.
It uses the graph of nearest neighbors to compute a higher-dimensional representation of the data, and then assigns labels using a *k*-means algorithm [PDSH]:

```python
from sklearn.cluster import SpectralClustering
model = SpectralClustering(n_clusters=2, affinity='nearest_neighbors',
                           assign_labels='kmeans')
labels = model.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis');
```

### Color compression

Lastly, let's see how *k*-means can be be used for compressing images.

One interesting application of clustering is in color compression within images. For example, imagine you have an image with millions of colors. In most images, a large number of the colors will be unused, and many of the pixels in the image will have similar or even identical colors [HML].

```python
from sklearn.datasets import load_sample_image
china = load_sample_image("china.jpg")
ax = plt.axes(xticks=[], yticks=[])
ax.imshow(china);
```

The image itself is stored in a three-dimensional array of size `(height, width, RGB)`, containing red/blue/green contributions as integers from 0 to 255 [PDSH]:

```python
china.shape
```

One way we can view this set of pixels is as a cloud of points in a three-dimensional color space.
We will reshape the data to `[n_samples, n_features]` and rescale the colors so that they lie between 0 and 1 [PDSH]:

```python
data = china / 255.0  # use 0...1 scale
data = data.reshape(-1, 3)
data.shape
```

We can visualize these pixels in this color space, using a subset of 10,000 pixels for efficiency (see the following figure) [PDSH]:

```python
def plot_pixels(data, title, colors=None, N=10000):
    if colors is None:
        colors = data

    # choose a random subset
    rng = np.random.default_rng(0)
    i = rng.permutation(data.shape[0])[:N]
    colors = colors[i]
    R, G, B = data[i].T

    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    ax[0].scatter(R, G, color=colors, marker='.')
    ax[0].set(xlabel='Red', ylabel='Green', xlim=(0, 1), ylim=(0, 1))

    ax[1].scatter(R, B, color=colors, marker='.')
    ax[1].set(xlabel='Red', ylabel='Blue', xlim=(0, 1), ylim=(0, 1))

    fig.suptitle(title, size=20);
```

```python
plot_pixels(data, title='Input color space: 16 million possible colors')
```

Now let's reduce these 16 million colors to just 16 colors, using a *k*-means clustering across the pixel space.
Because we are dealing with a very large dataset, we will use the mini-batch *k*-means, which operates on subsets of the data to compute the result (shown in the following figure) much more quickly than the standard *k*-means algorithm [PDSH]:

```python
from sklearn.cluster import MiniBatchKMeans
kmeans = MiniBatchKMeans(16, n_init='auto')
kmeans.fit(data)
new_colors = kmeans.cluster_centers_[kmeans.predict(data)]

plot_pixels(data, colors=new_colors,
            title="Reduced color space: 16 colors")
```

The result is a recoloring of the original pixels, where each pixel is assigned the color of its closest cluster center.
Plotting these new colors in the image space rather than the pixel space shows us the effect of this (see the following figure) [PDSH]:

```python
china_recolored = new_colors.reshape(china.shape)

fig, ax = plt.subplots(1, 2, figsize=(16, 6),
                       subplot_kw=dict(xticks=[], yticks=[]))
fig.subplots_adjust(wspace=0.05)
ax[0].imshow(china)
ax[0].set_title('Original Image', size=16)
ax[1].imshow(china_recolored)
ax[1].set_title('16-color Image', size=16);
```

Some detail is certainly lost in the rightmost panel, but the overall image is still easily recognizable.
In terms of the bytes required to store the raw data, the image on the right achieves a compression factor of around 1 million!
Now, this kind of approach is not going to match the fidelity of purpose-built image compression schemes like JPEG, but the example shows the power of thinking outside of the box with unsupervised methods like *k*-means [PDSH].
